```python
#import library
import sys

#put the output data into the file 
output=sys.stdout
outputfile=open("D:/CODEDATA/python/AY20_MBDS_questions/Question 1/output_question_1.txt","a") #“a”means append,"w"means rewrite
sys.stdout=outputfile

#input m,n,answear,and define the length
m=9
n=9
mn=m*n
length=m+n-1
answear=110


graph = {}

#Use "for" to construct a path that can be taken, that is, i  add 1 means to go right, or add n means to go down. 
    info = []
    if(i + 1 < mn):
        info.append(str(i+1))
    if(i + n < mn):
        info.append(str(i+n))
    graph[str(i)] = info

#define findAllPath function,Use recursive functions to find all paths 
def findAllPath(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return [path]

    paths = []  # Store all paths 
    for node in graph[start]:
        if node not in path:
            newpaths = findAllPath(graph, node, end, path)
            for newpath in newpaths: #For each path found, the sum judgment of the newpath and the length of the newpath are judged, and  if the path is matched, added to the path, else is directly eliminated. 
                sum=0
                for temp in newpath:
                    sum=sum+int(temp)//n+1
                if sum==answear and len(newpath)==length:
                    paths.append(newpath)
    return paths


#Call the function and get the result 
allpath = findAllPath(graph,'0','80')


#Convert the original 1-15 mark to RD 
final=[]
for path in allpath:
    temp=' '
    for i in range(len(path)-1):
        if(int(path[i+1])-int(path[i])==1):
            temp=temp+'R'
        elif(int(path[i+1])-int(path[i])==n):
            temp=temp+'D'
    final.append(temp)

print(answear," ",','.join(final) ) # but it is a pity that the second question has huge data , so my code could not run the result

'''
another solution ，though this solution cost more space,but it esay to understand

#import library
import itertools

#we can list all the possibility by using permutation
glist=set(itertools.permutations('RRRRRRRRDDDDDDDD',16))

answear=16
list1=[]

for i in glist:
    a=1
    sum=1
    for x in range(6):
        if i[x]=='R':
            sum=sum+a
        else:
            sum=sum+a+1
            a=a+1
    if sum==answear:
        list1.append(i)
print(set(list1))
   
'''
```
